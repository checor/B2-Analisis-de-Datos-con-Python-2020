```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/master/Datasets/diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
sns.histplot(df['glucose'])
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
sns.barplot(x=pregnancy_counts.index, y=pregnancy_counts)
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
sns.kdeplot(healthy['glucose'], fill=True)
sns.kdeplot(sick['glucose'], fill=True)
```

### d)

```python
sns.kdeplot(sick['blood_pressure'], fill=True)
sns.kdeplot(sick['glucose'], fill=True)
sns.kdeplot(sick['insulin'], fill=True)
```
